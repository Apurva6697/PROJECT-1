**EDA on Flight Fares in India ✈️**

This EDA analyzes flight prices across major cities of India. Routes we will be analyzing are:



* Mumbai(BOM) ↔️ Bengaluru(BLR)

* Delhi(DEL) ↔️ Bengaluru(BLR)
* Delhi(DEL) ↔️ Mumbai(BOM)


* Delhi(DEL) ↔️ Kolkata(CCU)

* Delhi(DEL) ↔️ Hyderabad (HYD)







Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

importing dataset

```python
df_1 = pd.read_csv('/content/flight_data_BOM_BLR.csv')
df_2 = pd.read_csv('/content/flight_data_DEL_BLR.csv')
df_3 = pd.read_csv('/content/flight_data_DEL_BOM.csv')
df_4 = pd.read_csv('/content/flight_data_DEL_CCU.csv')
df_5 = pd.read_csv('/content/flight_data_DEL_HYD.csv')
```

**Data Preprocessing**

```python
df_1.isnull().sum()
```

price column

```python
df_1["Price"] = df_1["Price"].str.replace(",","")
df_2["Price"] = df_2["Price"].str.replace(",","")
df_3["Price"] = df_3["Price"].str.replace(",","")
df_4["Price"] = df_4["Price"].str.replace(",","")
df_5["Price"] = df_5["Price"].str.replace(",","")
```

```python
df_1 = df_1.astype({"Price":"int"})
df_2 = df_2.astype({"Price":"int"})
df_3 = df_3.astype({"Price":"int"})
df_4 = df_4.astype({"Price":"int"})
df_5 = df_5.astype({"Price":"int"})
```

Departing Time

```python
df_1["DepartingHour"] = df_1["DepartingTime"].str.split(":").str[0]
df_2["DepartingHour"] = df_2["DepartingTime"].str.split(":").str[0]
df_3["DepartingHour"] = df_3["DepartingTime"].str.split(":").str[0]
df_4["DepartingHour"] = df_4["DepartingTime"].str.split(":").str[0]
df_5["DepartingHour"] = df_5["DepartingTime"].str.split(":").str[0]
```

```python
df_1 = df_1.astype({"DepartingHour":"int"})
df_2 = df_2.astype({"DepartingHour":"int"})
df_3 = df_3.astype({"DepartingHour":"int"})
df_4 = df_4.astype({"DepartingHour":"int"})
```

Average price per flight in route

```python
matplotlib.rcParams['figure.figsize'] = (20,10)
df_1_avg_price = df_1.groupby('FlightName')['Price'].mean().reset_index().astype({'Price':'int'}).sort_values(by='Price',ascending=False)
p = sns.barplot(x='FlightName',y='Price',data=df_1_avg_price)
p.bar_label(p.containers[0])
plt.title("Average Price per flight in route")
p.set_xticklabels(labels=df_1_avg_price["FlightName"], rotation=35)
sns.set(font_scale=1.25)
plt.show()
```

DepartingHour Vs Price

```python
df_dh_avg = df_1.groupby('DepartingHour')['Price'].mean().reset_index()
df_dh_avg.plot.line(x='DepartingHour',y='Price',title='DepartingHour Vs Price')
plt.show()
```

**Average price per route**

```python
df_avg_route = df.groupby(['DepartingCity','ArrivingCity'])['Price'].mean().reset_index().astype({'Price':'int'}).sort_values(by='Price',ascending=False)
df_avg_route["Route"] = df_avg_route["DepartingCity"]+"↔️"+df_avg_route["ArrivingCity"]
df_avg_route
```

```python
df_avg_route.plot.bar(x='Route',y='Price')
plt.show()
```
